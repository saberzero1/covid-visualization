```python
# Imports.
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import pandas as pd
import numpy as np
import plotly.offline as off
from plotly.subplots import make_subplots
from datetime import datetime
from IPython.display import display
```

```python
# Import datasets.
covid_df = pd.read_csv("Datasets/covid_19_india.csv")
vaccine_df = pd.read_csv("Datasets/covid_vaccine_statewise.csv")
```

```python
# Display cases dataset.
display(covid_df.head(10))
covid_df.info()
covid_df.describe()
```

```python
# Display vaccination dataset.
display(vaccine_df.head(10))
vaccine_df.info()
vaccine_df.describe()
```

```python
# Preprocessing datasets.
covid_df.drop(["Sno", "Time", "ConfirmedIndianNational", "ConfirmedForeignNational"], inplace = True, axis = 1)
```

```python
covid_df["Date"] = pd.to_datetime(covid_df["Date"], format = "%Y-%m-%d")
```

```python
# Active cases
covid_df["Active_Cases"] = covid_df["Confirmed"] - (covid_df["Cured"] + covid_df["Deaths"])
covid_df.tail()
```

```python
statewise = pd.pivot_table(covid_df, values = ["Confirmed", "Deaths", "Cured"], index = "State/UnionTerritory", aggfunc = max)
```

```python
# Recovery rate.
statewise["Recovery_Rate"] = statewise["Cured"] * 100 / statewise["Confirmed"]
```

```python
# Mortality rate.
statewise["Mortality_Rate"] = statewise["Deaths"] * 100 / statewise["Confirmed"]
```

```python
statewise = statewise.sort_values(by = "Confirmed", ascending = False)
```

```python
statewise.style.background_gradient(cmap = "BuPu")
```

```python
# Top 10 States by Active Cases.
top_ten_active_cases = covid_df.groupby(by = "State/UnionTerritory").max()[["Active_Cases", "Date"]].sort_values(by = ["Active_Cases"], ascending = False).reset_index()
# Create figure for Top 10 States by Active Cases.
fig = plt.figure(figsize=(16,9))
plt.title("Top 10 States by Active Cases", size = 25)
ax = sns.barplot(data = top_ten_active_cases.iloc[:10], y = "Active_Cases", x = "State/UnionTerritory", linewidth = 2, edgecolor = "black")
plt.xlabel("States")
plt.ylabel("Total Active Cases")
import plotly
import plotly.tools as tls
plotly_fig = tls.mpl_to_plotly(fig)
plotly.offline.plot(plotly_fig, filename="hi.html")
plt.show()
```

```python
# Top 10 States by Highest Deaths.
top_ten_deaths = covid_df.groupby(by = "State/UnionTerritory").max()[["Deaths", "Date"]].sort_values(by = ["Deaths"], ascending = False).reset_index()
# Create figure for Top 10 States by Active Cases.
fig = plt.figure(figsize=(18,5))
plt.title("Top 10 States by Total Deaths", size = 25)
ax = sns.barplot(data = top_ten_deaths.iloc[:10], y = "Deaths", x = "State/UnionTerritory", linewidth = 2, edgecolor = "black")
plt.xlabel("States")
plt.ylabel("Total Deaths")
fig.write_html("D:\Cloud\Repos\covid-visualization\Docs\Plots\total-deaths.html")
plt.show()
```

```python
# Top 5 States by Active Cases.
fig = plt.figure(figsize = (12,6))
ax = sns.lineplot(data = covid_df[covid_df["State/UnionTerritory"].isin(["Maharashtra", "Karnataka", "Kerala", "Tamil Nadu", "Uttar Pradesh"])], x = "Date", y = "Active_Cases", hue = "State/UnionTerritory")
ax.set_title("Top 5 Affected States", size = 16)
```

```python
vaccine_df.rename(columns = {"Updated On" : "Vaccination_Date"}, inplace = True)
vaccine_df.rename(columns = {"Total Individuals Vaccinated" : "Total"}, inplace = True)
```

```python
# Find missing values.
vaccine_df.isnull().sum()
```

```python
# Drop missing value columns.
vaccination = vaccine_df.drop(columns = ["Sputnik V (Doses Administered)", "AEFI", "18-44 Years (Doses Administered)", "45-60 Years (Doses Administered)", "60+ Years (Doses Administered)"], axis = 1)
```

```python
vaccination.head()
```

```python
# Vaccination rate of Males versus Females
male = vaccination["Male(Individuals Vaccinated)"].sum()
female = vaccination["Female(Individuals Vaccinated)"].sum()
px.pie(names = ["Male", "Female"], values = [male, female], title = "Male and Female Vaccination Rate", width=700, height=700)
```

```python
# Remove rows where stat is not India
vaccination = vaccination[vaccination["State"] != "India"]
vaccination
```

```python
# Most Vaccinated States.
max_vaccinated = vaccination.groupby("State")["Total"].sum().to_frame("Total").sort_values("Total", ascending = False)[:5]
max_vaccinated
```

```python
fix = plt.figure(figsize = (10,5))
plt.title("Top 5 Vaccinated States", size = 20)
x = sns.barplot(data = max_vaccinated.iloc[:10], y = max_vaccinated["Total"], x = max_vaccinated.index, linewidth = 2, edgecolor = "black")
plt.xlabel("States")
plt.ylabel("Vaccination")
plt.show()
```
